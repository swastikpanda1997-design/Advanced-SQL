# Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?

### Answer

A **Common Table Expression (CTE)** is a temporary named result set that exists only during the execution of a SQL query. It is created using the **WITH** keyword.

### Syntax

```sql
WITH CTE_Name AS
(
    SELECT ...
)
SELECT * FROM CTE_Name;
```

### Advantages

* Improves query readability.
* Breaks complex queries into smaller parts.
* Avoids repeating the same subquery.
* Makes SQL easier to debug and maintain.

# Q2. Why are some views updatable while others are read-only? Explain with an example.

### Answer

A **view** is a virtual table created from one or more tables.

### Updatable View

A view is **updatable** when:

* It is based on a single table.
* It does not contain GROUP BY, DISTINCT, Aggregate Functions, UNION, etc.

Example:

```sql
CREATE VIEW vw_Product
AS
SELECT ProductID, ProductName, Price
FROM Products;
```

Update using the view:

```sql
UPDATE vw_Product
SET Price = 1200
WHERE ProductID = 1;
```

---

### Read-only View

A view becomes read-only if it contains:

* GROUP BY
* DISTINCT
* SUM()
* AVG()
* COUNT()
* UNION
* JOIN (in many cases)

Example:

```sql
CREATE VIEW vw_Summary
AS
SELECT Category,
COUNT(*) TotalProducts
FROM Products
GROUP BY Category;
```

This view cannot be updated.

# Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?

### Answer

Stored Procedures provide several advantages:

* Reusable code
* Better performance (precompiled execution plan)
* Improved security
* Easier maintenance
* Reduces network traffic
* Supports parameters
* Centralizes business logic

Example:

```sql
CALL GetProducts('Electronics');
```

# Q4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.

### Answer

A **Trigger** is a special stored program that automatically executes when an event occurs.

Events include:

* INSERT
* UPDATE
* DELETE

### Uses

* Audit logging
* Data validation
* Automatic calculations
* Maintaining history

### Example

Whenever a product is deleted, automatically save it into an archive table.

# Q5. Explain the need for data modelling and normalization when designing a database.

### Answer

### Data Modelling

Data modelling defines how data is stored and how tables relate to each other.

Benefits:

* Better database design
* Easier querying
* Improved consistency
* Reduced redundancy

---

### Normalization

Normalization organizes data to reduce duplication.

Benefits:

* Eliminates duplicate data
* Improves data integrity
* Saves storage space
* Makes maintenance easier

Normal Forms:

* 1NF
* 2NF
* 3NF
* BCNF

# Q6. Write a CTE to calculate the total revenue for each product (Revenue = Price × Quantity), and return only products where revenue > 3000.

```sql
WITH ProductRevenue AS
(
    SELECT
        ProductID,
        ProductName,
        Price,
        Quantity,
        (Price * Quantity) AS Revenue
    FROM Products
)

SELECT *
FROM ProductRevenue
WHERE Revenue > 3000;
```

# Q7. Create a view named `vw_CategorySummary` that shows Category, TotalProducts, AveragePrice.

```sql
CREATE VIEW vw_CategorySummary AS
SELECT
    Category,
    COUNT(*) AS TotalProducts,
    AVG(Price) AS AveragePrice
FROM Products
GROUP BY Category;
```

View the data:

```sql
SELECT * FROM vw_CategorySummary;
```

# Q9. Create a stored procedure that accepts a category name and returns all products belonging to that category.

```sql
DELIMITER $$

CREATE PROCEDURE GetProductsByCategory
(
    IN p_Category VARCHAR(50)
)
BEGIN

SELECT *
FROM Products
WHERE Category = p_Category;

END$$

DELIMITER ;
```

Execute

```sql
CALL GetProductsByCategory('Electronics');
```

# Q10. Create an AFTER DELETE trigger that archives deleted product rows into `ProductArchive`.

### Create Archive Table

```sql
CREATE TABLE ProductArchive
(
    ProductID INT,
    ProductName VARCHAR(100),
    Category VARCHAR(50),
    Price DECIMAL(10,2),
    DeletedAt DATETIME
);
```

### Create Trigger

```sql
DELIMITER $$

CREATE TRIGGER trg_ProductDelete
AFTER DELETE
ON Products
FOR EACH ROW

BEGIN

INSERT INTO ProductArchive
(
    ProductID,
    ProductName,
    Category,
    Price,
    DeletedAt
)

VALUES
(
    OLD.ProductID,
    OLD.ProductName,
    OLD.Category,
    OLD.Price,
    NOW()
);

END$$

DELIMITER ;
```
